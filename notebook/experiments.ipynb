```python
from langchain.schema import Document
from langchain.vectorstores import FAISS
```

```python
# Create a Document object with the trimmed content
doc_chunk = Document(page_content=docs[0].page_content[:1000])

# Build the vectorstore
vectorstore = FAISS.from_documents([doc_chunk], embeddings_model)
```

## Question: user question
## Context: based on the question retrieving the info from the vector database

```python
prompt_template = """
        Answer the question based on the context provided below. 
        If the context does not contain sufficient information, respond with: 
        "I do not have enough information about this."

        Context: {context}

        Question: {question}

        Answer:"""
```

```python
from langchain.prompts import PromptTemplate
```

```python
prompt=PromptTemplate(
    template=prompt_template,
    input_variables=["context", "question"]
)
```

```python
prompt
```

```python
from langchain_core.output_parsers import StrOutputParser
```

```python
parser=StrOutputParser()
```

```python
def format_docs(docs):
    return "\n\n".join([doc.page_content for doc in docs])
```

```python
from langchain_core.runnables import RunnablePassthrough
```

```python
rag_chain = (
    {"context": retriever | format_docs, "question": RunnablePassthrough()}
    | prompt
    | model
    | StrOutputParser()
)
```

```python
rag_chain.invoke("tell  me about the llama2 finetuning benchmark experiments?")
```
